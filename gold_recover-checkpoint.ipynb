## Теория

Для общего понимания контекста задачи добавлю схематичное описание технологического процесса, его основные этапы и описание признаков. Так же, нам необходимо описать формулы расчета эффективности очистки золота и расчет метрики качества, задаваемой нам заказчиком. 

### Описание технологического процесса

![](https://pictures.s3.yandex.net/resources/viruchka_1576238830.jpg) 

Опишем каждую стадию: 
1. Флотация
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка 
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Описание данных


**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- - Xanthate — ксантогенат (промотер, или активатор флотации);
- - Sulphate — сульфат (на данном производстве сульфид натрия);
- - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота 

<br>**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи  

<br>**Наименование признаков**  
Наименование признаков должно быть такое:
- [этап].[тип_параметра].[название_параметра]    
Пример: rougher.input.feed_ag    

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики  

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно.   
Для этого вычислю её на обучающей выборке для признака rougher.output.recovery и найду MAE между расчётами и значением признака.
  
Эффективность обогащения рассчитывается по формуле
![формула эффективности](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)
где:
- C — доля золота в концентрате после флотации/очистки; rougher.output.concentrate_au
- F — доля золота в сырье/концентрате до флотации/очистки; rougher.input.feed_au
- T — доля золота в отвальных хвостах после флотации/очистки. rougher.output.tail_au

```python
recovery = data_train['rougher.output.recovery']
calculated_recovery = ((data_train['rougher.output.concentrate_au'] 
                      * (data_train['rougher.input.feed_au'] - data_train['rougher.output.tail_au'])) 
                     / (data_train['rougher.input.feed_au'] 
                        * (data_train['rougher.output.concentrate_au'] - data_train['rougher.output.tail_au']))  
                        * 100
                    )
print(mean_absolute_error(recovery, calculated_recovery))
```

Recovery для rougher расчитана верно, MAE между нашим расчетом и данными из датасета точны до -15 степени.

```python
def triangle_corr(data):
    #формируем матрицу корреляций
    corr_matrix = data.corr(method='spearman')
    #рисуем график
    plt.figure(figsize=(16, 16))
    np.triu(np.ones_like(corr_matrix))
    mask = np.triu(np.ones_like(corr_matrix, dtype=np.bool_))
    sns.heatmap(corr_matrix,mask=mask, vmin=-1, vmax=1, cmap='BrBG')
    plt.title('Треугольная тепловая карта матрицы корреляций', fontdict={'fontsize':18}, pad=16)
    plt.show()
    return

def triangle_corr_chaddock(data):
    #формируем матрицу корреляций
    corr_matrix = data.corr(method='spearman')
    new_corr = pd.DataFrame()
    for i in corr_matrix.columns:
        for j in corr_matrix.index:
            if np.abs(corr_matrix.loc[i,j]) >= CHADDOCK:
                new_corr.loc[i,j] = corr_matrix.loc[i,j]
            else:
                new_corr.loc[i,j] = 0
    #рисуем график
    plt.figure(figsize=(16, 16))
    np.triu(np.ones_like(new_corr))
    mask = np.triu(np.ones_like(new_corr, dtype=np.bool_))
    heatmap = sns.heatmap(new_corr,mask=mask, vmin=-1, vmax=1, cmap='BrBG')
    heatmap.set_title('Треугольная тепловая карта матрицы корреляций Чеддок', fontdict={'fontsize':18}, pad=16)
    #plt.title('Треугольная тепловая карта матрицы корреляций Чеддок, fontdict={'fontsize':18}, pad=16)
    plt.show()
    return new_corr
```

```python
def delete_corr_columns(data, corr_matrix):
    columns_for_del = []
    for i in corr_matrix.columns:
        for j in corr_matrix.index:
            if np.abs(corr_matrix.loc[i,j]) > CHADDOCK and corr_matrix.loc[i,j] < 1:
                columns_for_del.append(i)
                #data = data.remove(j, axis=1)
    columns_for_del = list(dict.fromkeys(columns_for_del))
    data = data.drop(columns_for_del, axis=1)
    print('Удалено колонок:', len(columns_for_del))
    return data
```

```python
def cut_over_std(data):
    # сохраняем описание целевых признаков
    describe_final = data['final.output.recovery'].describe()
    describe_rougher = data['rougher.output.recovery'].describe()
    # сохраняем 3 стандартных отклонения вправо и влево
    up_r = describe_rougher.loc['mean'] + describe_rougher.loc['std']*3
    down_r = describe_rougher.loc['mean'] - describe_rougher.loc['std']*3
    up_f = describe_final.loc['mean'] + describe_final.loc['std']*3
    down_f = describe_final.loc['mean'] - describe_final.loc['std']*3
    # обрезаем датасет
    data_new = data.loc[(data['rougher.output.recovery'] <= up_r) & (data['rougher.output.recovery'] >=down_r)]
    data_new = data_new.loc[(data_new['final.output.recovery'] <= up_f) & (data_new['final.output.recovery'] >=down_f)]
    return data_new
```

### Концентрация металлов (Au, Ag, Pb) на различных этапах очистки

Объявление переменных и функций, используемых в этом разделе и далее. 

```python
def find_element(row):    
    for i in elements:
        if i in row['index']:
            return i
    
def find_stage(row):
    for i in stages:
        if row['index'].find(i) == 0:
            return i
```

Для того, чтобы проанализировать концентрацию металлов, нам нужно отобрать признаки, характеризующие этот параметр.   
Мы можем собрать следующие этапы:
- rougher input
- rougher output
- primary cleaner
- final output  
Данные по содержанию в хвостах не берем, так как они нам по сути не интересны. 

```python
# выберем интересующие элементы и этапы
stages = ['rougher.input','rougher.output','primary_cleaner', 'final.output']
elements = ['au', 'ag', 'pb', 'sol']

element_columns = []
for column in list(data_full.columns):
    for element in elements:
        if element in column.split('_') and column.find('rougher.calculation.') == -1 and column.find('tail') == -1:
            element_columns.append(column)

#сразу рассортируем по очередности этапов
sorted_element_stage = []
for stage in stages:
    for row in element_columns:
        if stage in row:
            sorted_element_stage.append(row)
sorted_element_stage    
```

## Модель

Для начала заведем функцию для расчета метрики sMAPE. 

```python
def sMAPE(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    smape = 2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred))
    smape = smape.mean() * 100
    return smape
 
def result_smape(smape_rougher, result_final):
    result_smape = 0.25*smape_rougher + 0.75*result_final
    return result_smape

score = make_scorer(sMAPE, greater_is_better=False)
```

### Обучение модели

Обучим модель с помощью Линейной регрессии и метода Лассо. 
Используя функцию cross_val_score, применим кросс-валидацию, выделив 5 блоков, так как у нас нет валидационной выборки. 

#### Линейная регрессия:

```python
%%time
model_rougher_lr = LinearRegression()
smape_rougher_lr = -cross_val_score(model_rougher_lr, features_train, target_train_rougher, cv=3, scoring=score)

model_final_lr = LinearRegression()
smape_final_lr = -cross_val_score(model_final_lr, features_train, target_train_final, cv=3, scoring=score)

smape_res_lr = result_smape(smape_rougher_lr, smape_final_lr)
print('sMAPE для предсказания эффективности обогащения чернового концентрата rougher: {:.3f}'.format(smape_rougher_lr.mean()))
print('sMAPE для предсказания эффективности обогащения финального концентрата final: {:.3f}'.format(smape_final_lr.mean()))
print('\n Итоговый sMAPE: {:.3f}'.format(smape_res_lr.mean()))
```

#### Лассо

```python
%%time
model_lasso = Lasso(alpha=5)
cv = RepeatedKFold(n_splits=3, n_repeats=3, random_state=1)
smape_rougher_lasso = cross_val_score(model_lasso, features_train, target_train_rougher, scoring=score, cv=cv, n_jobs=-1)
scores_final_lasso = cross_val_score(model_lasso, features_train, target_train_final, scoring=score, cv=cv, n_jobs=-1)

smape_res_lasso = result_smape(smape_rougher_lasso, scores_final_lasso)

print('sMAPE для предсказания эффективности обогащения чернового концентрата rougher: {:.3f}'.format(-smape_rougher_lasso.mean()))
print('sMAPE для предсказания эффективности обогащения финального концентрата final: {:.3f}'.format(-scores_final_lasso.mean()))
print('\nИтоговый sMAPE  : {:.3f}'.format(-smape_res_lasso.mean()))
```

#### Случайный лес

```python
%%time
params = {'n_estimators': range(10,81,10), 'max_depth': range(2,11,2)} 

grf_r = GridSearchCV(RandomForestRegressor(random_state=42), 
                     param_grid=params, scoring=score, cv=3, n_jobs=-1)
grf_r.fit(features_train, target_train_rougher)

grf_f = GridSearchCV(RandomForestRegressor(random_state=42), 
                     param_grid=params, scoring=score, cv=3, n_jobs=-1)
grf_f.fit(features_train, target_train_final)

smape_res_f = result_smape(grf_r.best_score_ , grf_f.best_score_)
print('sMAPE для предсказания эффективности обогащения чернового концентрата rougher: {:.3f}. Params {}'
      .format(-grf_r.best_score_, grf_r.best_params_))
print('sMAPE для предсказания эффективности обогащения финального концентрата final: {:.3f}. Params {}'
      .format(-grf_f.best_score_, grf_f.best_params_))
print('\nИтоговый  sMAPE : {:.3f}\n'.format(-smape_res_f))
```
